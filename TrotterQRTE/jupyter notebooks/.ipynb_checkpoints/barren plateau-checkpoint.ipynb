```python
import numpy as np
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.circuit import Parameter, ParameterVector
from scipy.stats import gaussian_kde
from qiskit.quantum_info import SparsePauliOp, Statevector
# Parameters
num_qubits = 10
num_layers = 3
simulator = AerSimulator()

# Create a Parameterized Quantum Circuit
def create_ansatz(num_qubits, num_layers):
    circuit = QuantumCircuit(num_qubits)
    params = [Parameter(f'theta_{i}_{j}') for i in range(num_layers) for j in range(num_qubits)]
    param_index = 0

    for i in range(num_layers):
        for j in range(num_qubits):
            circuit.rx(params[param_index], j)
            param_index += 1
        for j in range(num_qubits - 1):
            circuit.cz(j, j+1)
    circuit.save_statevector()
    return circuit, params

# Create Cost Functions
# Global: <0|U|0> (State Fidelity)
def global_cost_function(statevector):
    return 1 - np.abs(statevector[0])**2

# Local: Measure each qubit in |0> state
def local_cost_function(statevector, num_qubits):
    probabilities = np.abs(statevector)**2
    expectation = 0
    for i in range(num_qubits):
        zero_state_prob = sum(probabilities[j] for j in range(len(probabilities)) if (j >> i) % 2 == 0)
        expectation += zero_state_prob / num_qubits
    return 1 - expectation

# Calculate Gradients using Parameter Shift Rule
def parameter_shift_gradient(circuit, params, cost_function):
    gradients = []
    shift = np.pi / 2
    for param_index, param in enumerate(params):
        # Positive shift
        param_values_plus = {p: v for p, v in zip(params, initial_values)}
        param_values_plus[param] += shift
        circuit_plus = circuit.assign_parameters(param_values_plus)
        result_plus = simulator.run(circuit_plus).result().get_statevector()
        cost_plus = cost_function(result_plus)

        # Negative shift
        param_values_minus = {p: v for p, v in zip(params, initial_values)}
        param_values_minus[param] -= shift
        circuit_minus = circuit.assign_parameters(param_values_minus)
        result_minus = simulator.run(circuit_minus).result().get_statevector()
        cost_minus = cost_function(result_minus)

        # Gradient calculation
        gradient = (cost_plus - cost_minus) / 2
        gradients.append(gradient)
    return gradients

# Run Experiment
circuit, params = create_ansatz(num_qubits, num_layers)
initial_values = np.random.uniform(0, 2 * np.pi, len(params))
bound_circuit = circuit.assign_parameters({p: v for p, v in zip(params, initial_values)})

# Evaluate Cost and Gradients
global_result = simulator.run(bound_circuit).result().get_statevector()
local_result = global_result.copy()

global_cost = global_cost_function(global_result)
local_cost = local_cost_function(local_result, num_qubits)

global_gradients = parameter_shift_gradient(circuit, params, global_cost_function)
local_gradients = parameter_shift_gradient(circuit, params, lambda s: local_cost_function(s, num_qubits))

# Plot Histogram
plt.figure(figsize=(10, 5))
plt.hist(global_gradients, bins=30, alpha=0.5, label='Global Cost Function', color='red')
plt.hist(local_gradients, bins=30, alpha=0.5, label='Local Cost Function', color='blue')
plt.xlabel('Gradient Values')
plt.ylabel('Frequency')
plt.title('Gradient Histogram: Global vs Local Cost Functions')
plt.legend()
plt.show()
```

```python
# Parameters
num_layers = 1
simulator = AerSimulator()
qubit_list = [2, 4, 6, 8, 10, 12]  # Number of qubits to test
num_trials = 100  # Repeat trials for averaging

# Tensor Product Ansatz with RX (matching paper)
def create_ansatz(num_qubits):
    circuit = QuantumCircuit(num_qubits)
    params = [Parameter(f'theta_{j}') for j in range(num_qubits)]
    for j in range(num_qubits):
        circuit.rx(params[j], j)
    circuit.save_statevector()
    return circuit, params

# Global Cost Function (target state is |0^n⟩)
def global_cost_function(statevector):
    statevector = np.asarray(statevector)
    zero_ket = np.zeros(len(statevector), dtype=complex)
    zero_ket[0] = 1.0
    fidelity = np.abs(np.vdot(zero_ket, statevector))**2
    return 1 - fidelity

# Local Cost Function: 1 - average probability of measuring 0 on each qubit
def local_cost_function(statevector, num_qubits):
    statevector = np.asarray(statevector)
    probabilities = np.abs(statevector)**2
    total = 0
    for i in range(num_qubits):
        prob_0 = sum(probabilities[j] for j in range(len(probabilities)) if (j >> i) % 2 == 0)
        total += prob_0
    return 1 - (total / num_qubits)

# Compute gradient using parameter-shift rule
def parameter_shift_gradient(circuit, params, initial_values, cost_function, cost_kwargs=None):
    gradients = []
    shift = np.pi / 2
    cost_kwargs = cost_kwargs or {}
    for param_index, param in enumerate(params):
        param_values_plus = {p: v for p, v in zip(params, initial_values)}
        param_values_plus[param] += shift
        circuit_plus = circuit.assign_parameters(param_values_plus)
        result_plus = simulator.run(circuit_plus).result().get_statevector()
        cost_plus = cost_function(result_plus, **cost_kwargs)

        param_values_minus = {p: v for p, v in zip(params, initial_values)}
        param_values_minus[param] -= shift
        circuit_minus = circuit.assign_parameters(param_values_minus)
        result_minus = simulator.run(circuit_minus).result().get_statevector()
        cost_minus = cost_function(result_minus, **cost_kwargs)

        gradient = (cost_plus - cost_minus) / 2
        gradients.append(gradient)
    return gradients

# Run experiment over different qubit sizes for global and local cost functions
global_variances = []
theoretical_variances = []
local_variances = []
theoretical_local_variances = []

for n in qubit_list:
    circuit, params = create_ansatz(n)

    var_trials_global = []
    var_trials_local = []
    for _ in range(num_trials):
        initial_values = np.random.uniform(0, 2 * np.pi, len(params))

        gradients_global = parameter_shift_gradient(
            circuit, params, initial_values, global_cost_function
        )
        var_trials_global.append(np.var(gradients_global))

        gradients_local = parameter_shift_gradient(
            circuit, params, initial_values, local_cost_function, cost_kwargs={"num_qubits": n}
        )
        var_trials_local.append(np.var(gradients_local))

    avg_var_global = np.mean(var_trials_global)
    avg_var_local = np.mean(var_trials_local)

    global_variances.append(avg_var_global)
    local_variances.append(avg_var_local)

    theoretical_global = (1/8)*(3/8)**(n-1)
    theoretical_local = 1 / (8 * n**2)
    theoretical_variances.append(theoretical_global)
    theoretical_local_variances.append(theoretical_local)

    print(f"Qubits: {n}, Global Var: {avg_var_global:.4e}, Local Var: {avg_var_local:.4e}, \
           Theoretical Global: {theoretical_global:.4e}, Theoretical Local: {theoretical_local:.4e}")

# Plot log-log variance vs number of qubits with both theoretical lines
plt.figure(figsize=(8, 5))
plt.plot(qubit_list, global_variances, 'o-', label='Global Cost Variance')
plt.plot(qubit_list, local_variances, 's-', label='Local Cost Variance')
plt.plot(qubit_list, theoretical_variances, 'd--', label='Theoretical Global $1/8 (3/8)^{n-1}$')
plt.plot(qubit_list, theoretical_local_variances, 'x--', label='Theoretical Local $1/8n^2$')
plt.yscale('log')
plt.xscale('log')
plt.xlabel('Number of Qubits (log)')
plt.ylabel('Gradient Variance (log)')
plt.title('Gradient Variance: Global vs Local Cost Functions')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Parameters
num_layers = 1
simulator = AerSimulator()
qubit_list = [4]  # Using only 4 qubits for comparison
num_trials = 30  # Repeat trials for averaging

# Alternating Layer Ansatz (Fig.3 of paper)
def create_ansatz(num_qubits, num_layers=1):
    circuit = QuantumCircuit(num_qubits)
    params = []
    for layer in range(num_layers):
        layer_params = [Parameter(f"theta_{layer}_{i}") for i in range(num_qubits)]
        params.extend(layer_params)
        for i in range(num_qubits):
            circuit.rx(layer_params[i], i)
        for i in range(num_qubits - 1):
            circuit.cx(i, i + 1)
    circuit.save_statevector()
    return circuit, params

# General expectation value: Tr[H |psi><psi|] = <psi|H|psi>
def hamiltonian_cost_function(statevector, hamiltonian):
    statevector = np.asarray(statevector)
    return np.real(Statevector(statevector).expectation_value(hamiltonian))

# Transverse-Field Ising Model Hamiltonian for 4 qubits
def tfim_hamiltonian(n, J=1.0, h=1.0):
    terms = []
    for i in range(n - 1):
        zz = ("I" * i) + "ZZ" + ("I" * (n - i - 2))
        terms.append((zz, -J))
    for i in range(n):
        x = ("I" * i) + "X" + ("I" * (n - i - 1))
        terms.append((x, -h))
    return SparsePauliOp.from_list(terms)

# Global projector Hamiltonian: H = I - |0^n><0^n|
def global_projector_hamiltonian(n):
    size = 2**n
    projector = np.zeros((size, size), dtype=complex)
    projector[0, 0] = 1.0
    identity = np.eye(size)
    return identity - projector

# Compute gradient using parameter-shift rule
def parameter_shift_gradient(circuit, params, initial_values, cost_function, cost_kwargs=None):
    gradients = []
    shift = np.pi / 2
    cost_kwargs = cost_kwargs or {}
    for param_index, param in enumerate(params):
        param_values_plus = {p: v for p, v in zip(params, initial_values)}
        param_values_plus[param] += shift
        circuit_plus = circuit.assign_parameters(param_values_plus)
        result_plus = simulator.run(circuit_plus).result().get_statevector()
        cost_plus = cost_function(result_plus, **cost_kwargs)

        param_values_minus = {p: v for p, v in zip(params, initial_values)}
        param_values_minus[param] -= shift
        circuit_minus = circuit.assign_parameters(param_values_minus)
        result_minus = simulator.run(circuit_minus).result().get_statevector()
        cost_minus = cost_function(result_minus, **cost_kwargs)

        gradient = (cost_plus - cost_minus) / 2
        gradients.append(gradient)
    return gradients

# Run experiment comparing TFIM and Global projector costs
qubit_list = [4]
tfim_variances = []
global_variances = []

for n in qubit_list:
    circuit, params = create_ansatz(n, num_layers)
    H_tfim = tfim_hamiltonian(n)
    H_global = global_projector_hamiltonian(n)

    var_trials_tfim = []
    var_trials_global = []

    for _ in range(num_trials):
        initial_values = np.random.uniform(0, 2 * np.pi, len(params))

        gradients_tfim = parameter_shift_gradient(
            circuit, params, initial_values, hamiltonian_cost_function,
            cost_kwargs={"hamiltonian": H_tfim}
        )
        var_trials_tfim.append(np.var(gradients_tfim))

        gradients_global = parameter_shift_gradient(
            circuit, params, initial_values, hamiltonian_cost_function,
            cost_kwargs={"hamiltonian": H_global}
        )
        var_trials_global.append(np.var(gradients_global))

    tfim_variances.append(np.mean(var_trials_tfim))
    global_variances.append(np.mean(var_trials_global))

    print(f"Qubits: {n}, TFIM Var: {tfim_variances[-1]:.4e}, Global Var: {global_variances[-1]:.4e}")

# Plot bar chart comparison
labels = [str(q) for q in qubit_list]
x = np.arange(len(labels))
width = 0.35

plt.figure(figsize=(6, 4))
plt.bar(x - width/2, tfim_variances, width, label='TFIM Cost')
plt.bar(x + width/2, global_variances, width, label='Global Projector Cost')
plt.xticks(x, labels)
plt.xlabel('Number of Qubits')
plt.ylabel('Gradient Variance')
plt.title('TFIM vs Global Projector Cost (n=4)')
plt.legend()
plt.tight_layout()
plt.show()

```

```python
from qiskit.quantum_info import SparsePauliOp, Pauli
import numpy as np
import random
from itertools import combinations
del list
# Extract support submatrix and compute trace and Tr(O^2) only on non-I part
def support_trace_info(label):
    support = [c for c in label if c != 'I']
    support_str = ''.join(support)
    if support_str == '':
        tr_O = 2 ** len(label)
        tr_O2 = tr_O  # identity
    else:
        pauli_matrix = Pauli(support_str).to_matrix()
        tr_O = np.trace(pauli_matrix)
        tr_O2 = np.trace(pauli_matrix @ pauli_matrix.conj().T)
    return tr_O, tr_O2

def analyze_case_ii_structure(pauli_op: SparsePauliOp):
    terms = pauli_op.to_list()
    results = []
    for label, coeff in terms:
        non_identity_count = sum(1 for c in label if c != 'I')

        full_matrix = Pauli(label).to_matrix()
        tr_full = np.trace(full_matrix)

        tr_support, tr_support2 = support_trace_info(label)

        term_info = {
            "pauli_string": label,
            "non_identity_count": non_identity_count,
            "trace_full": tr_full,
            "trace_support": tr_support,
            "is_traceless_support": np.isclose(tr_support, 0.0),
            "tr_O2_support": tr_support2.real,
            "expected_2^m": 2 ** non_identity_count
        }
        results.append(term_info)
    return results

# Construct local cost observable as in Eq. (16): O = sum_i c_i A_i B_i
# Each A_i and B_i acts on m/2 qubits, total support m <= n

def generate_local_cost_operator(n=6, m=4, num_terms=5):
    paulis = ['I', 'X', 'Y', 'Z']
    terms = []
    for _ in range(num_terms):
        # randomly choose two disjoint subsets of size m//2
        all_qubits = list(range(n))
        support_A = sorted(random.sample(all_qubits, m // 2))
        support_B = sorted(random.sample([q for q in all_qubits if q not in support_A], m // 2))

        # construct Pauli strings for A and B
        A = ['I'] * n
        B = ['I'] * n
        for q in support_A:
            A[q] = random.choice(paulis[1:])  # non-identity
        for q in support_B:
            B[q] = random.choice(paulis[1:])

        AB = [a if a != 'I' else b for a, b in zip(A, B)]
        # If both A and B are non-I at the same position, multiply them
        for i in range(n):
            if A[i] != 'I' and B[i] != 'I' and A[i] != B[i]:
                AB[i] = 'Z'  # simplification: not full multiplication
        terms.append((''.join(AB), 1.0))

    return SparsePauliOp.from_list(terms)

# Generate and analyze a local cost operator
H_local = generate_local_cost_operator(n=6, m=4, num_terms=5)
analysis = analyze_case_ii_structure(H_local)

print("Local cost operator (Eq. 16-style) analysis:")
for i, term in enumerate(analysis):
    print(f"Term {i+1}: {term['pauli_string']}, non-I count = {term['non_identity_count']}, Tr(support) = {term['trace_support']:.2f}, Tr(O^2 support) = {term['tr_O2_support']:.2f}, Expected 2^m = {term['expected_2^m']}, Traceless: {term['is_traceless_support']}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit
from qiskit_aer.noise import NoiseModel, depolarizing_error
from qiskit.quantum_info import SparsePauliOp
from qiskit_aer.primitives import EstimatorV2

def ansatz_hea(nqubits, depth, params):
    qc = QuantumCircuit(nqubits)
    idx = 0
    for d in range(depth):
        for i in range(nqubits):
            qc.ry(params[idx], i)
            idx += 1
        for i in range(nqubits - 1):
            qc.cx(i, i + 1)
    return qc

def get_hamiltonian(n):
    h_terms = []
    for i in range(n - 1):
        # Build label like "ZZII" for Z_i Z_{i+1}
        label = ['I'] * n
        label[i] = 'Z'
        label[i + 1] = 'Z'
        h_terms.append((''.join(label), -1.0))
    for i in range(n):
        label = ['I'] * n
        label[i] = 'X'
        h_terms.append((''.join(label), -1.0))
    return SparsePauliOp.from_list(h_terms)


def build_noise_model(error_rate):
    noise_model = NoiseModel()
    depol1 = depolarizing_error(error_rate, 1)
    depol2 = depolarizing_error(error_rate * 10, 2)
    noise_model.add_all_qubit_quantum_error(depol1, ['u1', 'u2', 'u3'])
    noise_model.add_all_qubit_quantum_error(depol2, ['cx'])
    return noise_model

def cost_function(params, nqubits, depth, error_rate):
    hamiltonian = get_hamiltonian(nqubits)
    circuit = ansatz_hea(nqubits, depth, params)
    circuit = circuit.decompose()
    if error_rate == 0:
        estimator = EstimatorV2()
    else:
        noise_model = build_noise_model(error_rate)
        estimator = EstimatorV2(options={"backend_options": {"noise_model": noise_model}})
    result = estimator.run([(circuit, hamiltonian)]).result()
    return result[0].data.evs

def compute_gradient(params, nqubits, depth, error_rate, eps=1e-2):
    grads = []
    for i in range(len(params)):
        p_plus = params.copy()
        p_minus = params.copy()
        p_plus[i] += eps
        p_minus[i] -= eps
        grad = (cost_function(p_plus, nqubits, depth, error_rate) -
                cost_function(p_minus, nqubits, depth, error_rate)) / (2 * eps)
        grads.append(grad)
    return np.array(grads)

def run_experiment(nqubits, depth, error_rate, trials=10):
    vars = []
    for _ in range(trials):
        params = 2 * np.pi * np.random.rand(nqubits * depth)
        grad = compute_gradient(params, nqubits, depth, error_rate)
        vars.append(np.var(grad))
    return np.mean(vars)

# Run experiments
nqubits_list = [2, 4, 6, 8, 10]
depth = 1
error_rates = [0.0, 0.1]
results = {e: [] for e in error_rates}

for n in nqubits_list:
    for e in error_rates:
        var = run_experiment(n, depth, e, trials=5)
        results[e].append(var)

# Plot results
for e in error_rates:
    plt.plot(nqubits_list, results[e], label=f"error rate = {e}")
plt.xlabel("Number of Qubits")
plt.ylabel("Gradient Variance (log scale)")
plt.yscale("log")
plt.title("Gradient Variance under Noise")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector, SparsePauliOp
from matplotlib.lines import Line2D
import pandas as pd

def build_ansatz(thetas, nqubits):
    qc = QuantumCircuit(nqubits)
    idx = 0
    for i in range(nqubits):
        qc.rx(thetas[idx], i)
        idx += 1
        qc.ry(thetas[idx], i)
        idx += 1
    return qc

# 构建 global Hamiltonian：Z ⊗ Z ⊗ Z ...
def get_global_H(nqubits):
    return SparsePauliOp("Z" * nqubits, coeffs=[1.0])

# 构建 local Hamiltonian：TFIM = -sum Z_i Z_{i+1} - sum X_i
def get_local_H(nqubits):
    terms = []
    for i in range(nqubits - 1):
        label = ["I"] * nqubits
        label[i] = "Z"
        label[i + 1] = "Z"
        terms.append(("".join(label), -1.0))
    for i in range(nqubits):
        label = ["I"] * nqubits
        label[i] = "X"
        terms.append(("".join(label), -1.0))
    return SparsePauliOp.from_list(terms)

# 计算 expectation value
def expectation(thetas, nqubits, H):
    qc = build_ansatz(thetas, nqubits)
    state = Statevector.from_instruction(qc)
    return np.real(state.expectation_value(H))

# 构造 surface，扫描两个参数，其余随机（Haar）
def generate_surface_expectation(nqubits, H, x_idx, y_idx, step=0.6):
    X, Y = np.meshgrid(np.arange(-np.pi, np.pi, step),
                       np.arange(-np.pi, np.pi, step))
    Z = np.zeros_like(X)
    total_params = 2 * nqubits
    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            thetas = np.random.uniform(-np.pi, np.pi, size=total_params)
            thetas[x_idx] = X[i, j]
            thetas[y_idx] = Y[i, j]
            Z[i, j] = expectation(thetas, nqubits, H)
    return X, Y, Z

# 画图
def plot_overlayed_surfaces(X, Y, Z1, Z2, label1, label2, title):
    fig = plt.figure(figsize=(10, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z1, color='royalblue', alpha=0.8)
    ax.plot_surface(X, Y, Z2, color='orangered', alpha=0.6)
    ax.set_title(title)
    ax.set_xlabel(r'$\theta_2$')
    ax.set_ylabel(r'$\theta_5$')
    ax.set_zlabel("Expectation")
    custom_lines = [
        Line2D([0], [0], color='royalblue', lw=4),
        Line2D([0], [0], color='orangered', lw=4)
    ]
    ax.legend(custom_lines, [label1, label2])
    ax.view_init(elev=45, azim=-60)
    plt.tight_layout()
    plt.show()

# 运行实验：global -> Z...Z； local -> TFIM
x_idx, y_idx = 2, 5
X4, Y4, Zg4 = generate_surface_expectation(4, get_global_H(4), x_idx, y_idx)
_, _, Zg10 = generate_surface_expectation(10, get_global_H(10), x_idx, y_idx)
#plot_overlayed_surfaces(X4, Y4, Zg4, Zg10, "Global Z...Z (n=4)", "Global Z...Z (n=10)", "Global Expectation Comparison")

X4, Y4, Zl4 = generate_surface_expectation(4, get_local_H(4), x_idx, y_idx)
_, _, Zl10 = generate_surface_expectation(10, get_local_H(10), x_idx, y_idx)
#plot_overlayed_surfaces(X4, Y4, Zl4, Zl10, "TFIM (n=4)", "TFIM (n=10)", "Local TFIM Expectation Comparison")
def save_surface_data(X, Y, Z, filename):
    flat_data = {
        "theta2": X.flatten(),
        "theta5": Y.flatten(),
        "expectation": Z.flatten()
    }
    df = pd.DataFrame(flat_data)
    df.to_csv(filename, sep='\t', index=False)

# 保存四组数据
save_surface_data(X4, Y4, Zg4, "global_expectation_n4.txt")
save_surface_data(X4, Y4, Zg10, "global_expectation_n10.txt")
save_surface_data(X4, Y4, Zl4, "local_expectation_n4.txt")
save_surface_data(X4, Y4, Zl10, "local_expectation_n10.txt")
```

```python
# Re-run the gradient experiment code after kernel reset

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit, transpile
from qiskit.quantum_info import SparsePauliOp
from qiskit_aer import AerSimulator
from qiskit_aer.noise import NoiseModel, depolarizing_error

# Build RX-RY ansatz with depth
def build_ansatz(thetas, nqubits, depth):
    qc = QuantumCircuit(nqubits)
    idx = 0
    for d in range(depth):
        for i in range(nqubits):
            qc.ry(thetas[idx], i)
            idx += 1
            qc.rx(thetas[idx], i)
            idx += 1
    return qc

# Global Hamiltonian: Z...Z
def get_global_H(nqubits):
    return SparsePauliOp("Z" * nqubits, coeffs=[1.0])

# Local (TFIM-like): -Z_i Z_{i+1} - X_i
def get_local_H(nqubits):
    terms = []
    for i in range(nqubits - 1):
        label = ['I'] * nqubits
        label[i], label[i + 1] = 'Z', 'Z'
        terms.append(("".join(label), -1.0))
    for i in range(nqubits):
        label = ['I'] * nqubits
        label[i] = 'X'
        terms.append(("".join(label), -1.0))
    return SparsePauliOp.from_list(terms)

# Build noise model (depolarizing)
def build_noise_model(error_rate):
    noise_model = NoiseModel()
    noise_model.add_all_qubit_quantum_error(depolarizing_error(error_rate, 1), ['rx', 'ry', 'id'])
    noise_model.add_all_qubit_quantum_error(depolarizing_error(error_rate * 10, 2), ['cx'])
    return noise_model

# Compute expectation from density matrix simulation
def expectation_density_matrix(thetas, nqubits, depth, H, simulator):
    qc = build_ansatz(thetas, nqubits, depth)
    qc.save_density_matrix()
    tqc = transpile(qc, simulator)
    result = simulator.run(tqc).result()
    rho = result.data(0)['density_matrix']
    return np.real(rho.expectation_value(H))

# Compute gradient using parameter-shift rule (numerical)
def compute_gradient(thetas, nqubits, depth, H, simulator, target_idx, eps=1e-2):
    shifted_plus = thetas.copy()
    shifted_minus = thetas.copy()
    shifted_plus[target_idx] += eps
    shifted_minus[target_idx] -= eps
    f_plus = expectation_density_matrix(shifted_plus, nqubits, depth, H, simulator)
    f_minus = expectation_density_matrix(shifted_minus, nqubits, depth, H, simulator)
    return (f_plus - f_minus) / (2 * eps)

# Run experiment for global/local cost over different qubit counts
def run_gradient_experiment(nqubits_list, depth, error_rate):
    grads_global = []
    grads_local = []
    for n in nqubits_list:
        total_params = 2 * n * depth
        thetas = np.random.uniform(-np.pi, np.pi, total_params)
        last_layer_start = 2 * n * (depth - 1)
        target_idx = last_layer_start + 1  # RX on first qubit in last depth layer

        noise_model = build_noise_model(error_rate)
        simulator = AerSimulator(method='density_matrix', noise_model=noise_model)

        grad_g = compute_gradient(thetas, n, depth, get_global_H(n), simulator, target_idx)
        grad_l = compute_gradient(thetas, n, depth, get_local_H(n), simulator, target_idx)

        grads_global.append(grad_g)
        grads_local.append(grad_l)

    return grads_global, grads_local

# Define parameters
nqubits_list = [2, 4, 8, 10]
depth = 2
error_rate= 0.1
# Run gradient experiments
grads_global, grads_local = run_gradient_experiment(nqubits_list, depth, error_rate)

# Prepare and show results
df = pd.DataFrame({
    'nqubits': nqubits_list,
    'Gradient (Global)': grads_global,
    'Gradient (Local)': grads_local
})

df.set_index('nqubits').plot(marker='o', logy=False)
plt.xlabel('Number of Qubits')
plt.ylabel('Gradient (∂E/∂θ)')
plt.title(f'Gradient w.r.t. last RX parameter (noise parameter{error_rate}, depth=2)')
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit, transpile
from qiskit.quantum_info import SparsePauliOp
from qiskit_aer import AerSimulator
from qiskit_aer.noise import NoiseModel, depolarizing_error

# 1. 构造 ansatz
def build_ansatz(thetas, nqubits, depth):
    qc = QuantumCircuit(nqubits)
    idx = 0
    for d in range(depth):
        for i in range(nqubits):
            qc.ry(thetas[idx], i)
            idx += 1
            qc.rx(thetas[idx], i)
            idx += 1
    return qc
def ansatz_hamiltonian(params, nqubits, depth):
    circuit = QuantumCircuit(nqubits)
    t = 0
    for i in range(nqubits):
        circuit.h(i)
    for idepth in range(depth):
        for i in range(nqubits):
            if i < (nqubits - 1):
                j = i + 1
            else:
                j = 0
            circuit.rzz(params[t], i, j)
            t += 1
        circuit.barrier()
        for i in range(nqubits):
            if i < (nqubits - 1):
                j = i + 1
            else:
                j = 0
            circuit.rx(params[t], i)
            t += 1
    return circuit
# 2. 构造 Hamiltonians
def get_global_H(nqubits):
    return SparsePauliOp("Z" * nqubits, coeffs=[1.0])

def get_local_H(nqubits):
    terms = []
    for i in range(nqubits - 1):
        label = ['I'] * nqubits
        label[i], label[i + 1] = 'Z', 'Z'
        terms.append(("".join(label), -1.0))
    for i in range(nqubits):
        label = ['I'] * nqubits
        label[i] = 'X'
        terms.append(("".join(label), -1.0))
    return SparsePauliOp.from_list(terms)

# 3. 噪声模型
def build_noise_model(error_rate):
    noise_model = NoiseModel()
    noise_model.add_all_qubit_quantum_error(depolarizing_error(error_rate, 1), ['rx', 'ry', 'id'])
    noise_model.add_all_qubit_quantum_error(depolarizing_error(error_rate * 10, 2), ['cx'])
    return noise_model

# 4. 期望值计算
def expectation_density_matrix(thetas, nqubits, depth, H, simulator):
    qc = build_ansatz(thetas, nqubits, depth)
    qc.save_density_matrix()
    tqc = transpile(qc, simulator)
    result = simulator.run(tqc).result()
    rho = result.data(0)['density_matrix']
    return np.real(rho.expectation_value(H))

# 5. 梯度计算（有限差分）
def compute_gradient(thetas, nqubits, depth, H, simulator, target_idx, eps=1e-2):
    shifted_plus = thetas.copy()
    shifted_minus = thetas.copy()
    shifted_plus[target_idx] += eps
    shifted_minus[target_idx] -= eps
    f_plus = expectation_density_matrix(shifted_plus, nqubits, depth, H, simulator)
    f_minus = expectation_density_matrix(shifted_minus, nqubits, depth, H, simulator)
    return (f_plus - f_minus) / (2 * eps)

# 6. 主实验：计算每个 nqubits 下的梯度均值和方差
def run_gradient_stats(nqubits_list, error_rate, trials=30, repeats=1):
    stats = []
    for n in nqubits_list:
        depth = n - 1
        total_params = 2 * n * depth
        last_layer_start = 2 * n * (depth - 1)
        target_idx = last_layer_start + 1  # 最后一层第一个 qubit 上的 RX

        noise_model = build_noise_model(error_rate)
        simulator = AerSimulator(method='density_matrix', noise_model=noise_model)

        grads_g, grads_l = [], []
        for _ in range(trials):
            thetas = np.random.uniform(-np.pi, np.pi, total_params)

            #
            grad_g = np.mean([
                compute_gradient(thetas, n, depth, get_global_H(n), simulator, target_idx)
                for _ in range(repeats)
            ])
            grad_l = np.mean([
                compute_gradient(thetas, n, depth, get_local_H(n), simulator, target_idx)
                for _ in range(repeats)
            ])

            grads_g.append(grad_g)
            grads_l.append(grad_l)

        stats.append({
            "nqubits": n,
            "mean_grad_global": np.mean(grads_g),
            "var_grad_global": np.var(grads_g),
            "mean_grad_local": np.mean(grads_l),
            "var_grad_local": np.var(grads_l)
        })

    return pd.DataFrame(stats)


# 7. 执行实验
nqubits_list = [2, 4, 6, 8, 10, 12]
error_rate = 0.1
df = run_gradient_stats(nqubits_list, error_rate)

# 8. 作图：梯度方差
plt.figure(figsize=(8, 5))
plt.plot(df['nqubits'], df['var_grad_global'], 'o-', label='Global')
plt.plot(df['nqubits'], df['var_grad_local'], 's--', label='Local')
plt.yscale('log')
plt.xlabel("Number of Qubits")
plt.ylabel("Gradient Variance (log scale)")
plt.title(f"Variance of Gradient w.r.t. RX in last layer (noise parameter {error_rate})")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from generate_surface_from_ansatz import generate_surface_from_ansatz
from qiskit.quantum_info import SparsePauliOp
from qiskit_aer import AerSimulator
from qiskit_aer.noise import NoiseModel, depolarizing_error
from qiskit import QuantumCircuit, transpile
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import LinearLocator, FormatStrFormatter

# ==== 可选 ansatz ====
def ansatz_hva(params, nqubits, depth):
    qc = QuantumCircuit(nqubits)
    t = 0
    for i in range(nqubits):
        qc.h(i)
    for _ in range(depth):
        for i in range(nqubits):
            j = (i + 1) % nqubits
            qc.rzz(params[t], i, j)
            t += 1
        for i in range(nqubits):
            qc.rx(params[t], i)
            t += 1
    return qc

def ansatz_hea(params, nqubits, depth):
    qc = QuantumCircuit(nqubits)
    t = 0
    for _ in range(depth):
        for i in range(nqubits):
            qc.ry(params[t], i)
            t += 1
        for i in range(nqubits - 1):
            qc.cx(i, i + 1)
    return qc

# ==== Hamiltonians ====
def get_global_H(nqubits):
    return SparsePauliOp("Z" * nqubits, coeffs=[1.0])

def get_local_H(nqubits):
    terms = []
    for i in range(nqubits - 1):
        label = ['I'] * nqubits
        label[i], label[i + 1] = 'Z', 'Z'
        terms.append(("".join(label), -1.0))
    for i in range(nqubits):
        label = ['I'] * nqubits
        label[i] = 'X'
        terms.append(("".join(label), -1.0))
    return SparsePauliOp.from_list(terms)

# ==== Noise ====
def build_noise_model(error_rate):
    noise_model = NoiseModel()
    noise_model.add_all_qubit_quantum_error(depolarizing_error(error_rate, 1), ['rx', 'ry', 'id'])
    noise_model.add_all_qubit_quantum_error(depolarizing_error(error_rate * 10, 2), ['cx', 'rzz'])
    return noise_model

def get_simulator(error_rate):
    return AerSimulator(method='density_matrix', noise_model=build_noise_model(error_rate))

# ==== 绘图 ====
def plot_and_save_surface(X, Y, Z, title, filename):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection="3d")
    ax.plot_surface(X, Y, Z, cmap="viridis", linewidth=0, antialiased=False)
    ax.set_title(title)
    ax.set_xlabel("θ_x")
    ax.set_ylabel("θ_y")
    ax.set_zlabel("Cost")
    ax.set_zlim(-4, 4)
    ax.zaxis.set_major_locator(LinearLocator(10))
    ax.zaxis.set_major_formatter(FormatStrFormatter("%.02f"))
    plt.tight_layout()
    plt.savefig(filename)
    plt.show()

# ==== 主程序 ====
if __name__ == "__main__":
    nqubits = 4
    depth = nqubits - 1
    error_rate = 0.1  # or 0.0 for ideal
    ansatz_type = "hva"  # or "hea"
    H = get_local_H(nqubits)  # or get_global_H(nqubits)

    # 选择两个变量参数的索引
    if ansatz_type == "hea":
        total_params = nqubits * depth
        x_idx, y_idx = total_params - 2, total_params - 1
    else:
        total_params = 2 * nqubits * depth
        x_idx, y_idx = total_params - 2, total_params - 1

    # 生成 landscape
    X, Y, Z = generate_surface_from_ansatz(nqubits, depth, H, error_rate, ansatz_type, x_idx, y_idx)

    # 保存和显示图像
    title = f"{ansatz_type.upper()} Cost Landscape (n={nqubits}, error={error_rate})"
    filename = f"landscape_{ansatz_type}_n{nqubits}_error{error_rate}.png"
    plot_and_save_surface(X, Y, Z, title, filename)
```
