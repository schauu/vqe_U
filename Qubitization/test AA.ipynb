```python
import numpy as np
import pennylane as qml
from scipy.linalg import expm
import math
```

```python
num_wires = 2
dev = qml.device("default.qubit", wires=num_wires)

@qml.qnode(dev)
def circuit():
    for wire in range(num_wires):
        qml.Hadamard(wire)
    qml.FlipSign([1, 0], wires=range(num_wires))
    return qml.state()

circuit()
```

```python
num_wires = 2
dev = qml.device("default.qubit", wires=num_wires)

@qml.qnode(dev)
def circuit():
    for wire in range(num_wires):
        qml.Hadamard(wire)
    qml.FlipSign(n=0, wires=[0])
    return qml.state()

circuit()
```
